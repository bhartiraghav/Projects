```python
import sqlite3
from tkinter import *
from tkinter.ttk import Combobox
from tkinter import messagebox
from datetime import datetime
import time
from tkinter.ttk import Style,Treeview,Scrollbar
```

```python
try: 
    conobj=sqlite3.connect(database="banking.sqlite")
    curobj=conobj.cursor()
    curobj.execute("create table accounts(acn integer primary key autoincrement,name text,email text,mob text,pswd text,c_pswd text,bal float,opdate text)")
    curobj.execute("create table txns(acn int,amt float,updatebal float,type text,txndate text)")      
    curobj.execute("create table query(acn int,name text,query text)")
    conobj.close()
    print("table created")
except:
    print("something went wrong, maybe table is already exists")
```

```python

win=Tk()
win.configure(bg="powder blue")
win.resizable(height=False,width=False)
win.state("zoom")

title=Label(win,text="Banking Automataion",font=("arial",30,"bold","underline"),bg="powder blue")
title.pack()

date=Label(win,text=f"{datetime.now().date()}",font=("arial",10,"bold"),bg="powder blue")
date.pack()

def mainscreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.1,relwidth=1,relheight=.9)
    frm.configure(bg="powder blue")
    
    def new():
        frm.destroy()
        newaccscreen()
        
    def fg():
        frm.destroy()
        fgscreen()
    
    def reset():
        e_acn.delete(0,"end")
        e_pws.delete(0,"end")
        e_acn.focus()
        
    def qhelp():
        frm.destroy()
        queryscreen()
        
    def log():
        acn=e_acn.get()
        pswd=e_pws.get()
        
        if (len(pswd)==0 and len(acn)==0):
            messagebox.showerror("login","empty fields not allowed")
        else:
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select * from accounts where acn=? and pswd=?",(acn,pswd))
            tup=curobj.fetchone()
            
            if tup==None:
                messagebox.showerror("lodin failed","invailed ACN/pswd")
            else:
                global uname,uacn
                uacn=tup[0]
                uname=tup[1]
                frm.destroy()
                homescreen()
    
    lbl_acn=Label(frm,text="Account No.",font=("arial",20),bg="powder blue")
    lbl_acn.place(relx=.3,rely=.2)
    
    e_acn=Entry(frm,font=("arial",20),bd=2,width=16)
    e_acn.place(relx=.45,rely=.2)
    
    lbl_pws=Label(frm,text="Password",font=("arial",20),bg="powder blue")
    lbl_pws.place(relx=.3,rely=.3)
    
    e_pws=Entry(frm,font=("arial",20),bd=2,width=16,show="*")
    e_pws.place(relx=.45,rely=.3)
    
    bt_lg=Button(frm,text="Login",font=("arial",20),command=log)
    bt_lg.place(relx=.47,rely=.4)

    bt_reset=Button(frm,text="Reset",font=("arial",20),command=reset)
    bt_reset.place(relx=.55,rely=.4)
    
    bt_fg=Button(frm,text="Forgot password",font=("arial",20),command=fg)
    bt_fg.place(relx=.1,rely=.6)
    
    bt_new=Button(frm,text="Create Account ",font=("arial",20),command=new)
    bt_new.place(relx=.7,rely=.6)
    
    
    bt_qr=Button(frm,text="Queries ",font=("arial",20),command=qhelp)
    bt_qr.place(relx=.43,rely=.6)
    

def newaccscreen():
    frm=Frame(win)
    frm.configure(bg="powder blue")
    frm.place(relx=0,rely=.1,relwidth=1,relheight=.9)
    
    def back():
        frm.destroy()
        mainscreen()
        
    def openacdb():
        bal=0
        name=e_name.get()
        email=e_eml.get()
        mob=e_ph.get()
        actype=cb_type.get()
        opdate=time.ctime()
        pswd=e_pws.get()
        cpswd=e_cpws.get()
        
        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("insert into accounts(name,email,mob,pswd,c_pswd,bal,opdate,type) values(?,?,?,?,?,?,?,?)",(name,email,mob,pswd,cpswd,bal,opdate,actype))
        conobj.commit()
        curobj.close()
        
        curobj=conobj.cursor()
        curobj.execute("select max(acn) from accounts")
        tup=curobj.fetchone()
        messagebox.showinfo("account",f"account opend with Acc no. {tup[0]}")
        conobj.close()
        
    
    lbl_name=Label(frm,text="Name",font=("arial",20),bg="powder blue")
    lbl_name.place(relx=.33,rely=.2)
    
    e_name=Entry(frm,font=("arial",20),bd=2,width=16)
    e_name.place(relx=.5,rely=.2)
    
    lbl_eml=Label(frm,text="Email",font=("arial",20),bg="powder blue")
    lbl_eml.place(relx=.33,rely=.3)
    
    e_eml=Entry(frm,font=("arial",20),bd=2,width=16)
    e_eml.place(relx=.5,rely=.3)
    
            
    lbl_ph=Label(frm,text="mobile No.",font=("arial",20),bg="powder blue")
    lbl_ph.place(relx=.29,rely=.4)
    
    e_ph=Entry(frm,font=("arial",20),bd=2,width=16)
    e_ph.place(relx=.5,rely=.4)
    
    lbl_ac=Label(frm,text="Acc Type",font=("arial",20),bg="powder blue")
    lbl_ac.place(relx=.3,rely=.5)
    
    cb_type=Combobox(frm,values=["saving","current"],font=("arial",20),width=15)
    cb_type.current(0)
    cb_type.place(relx=.5,rely=.5)
    

    lbl_pws=Label(frm,text="password",font=("arial",20),bg="powder blue")
    lbl_pws.place(relx=.3,rely=.6)
    
    e_pws=Entry(frm,font=("arial",20),bd=2,width=16,show="*")
    e_pws.place(relx=.5,rely=.6)
    
    lbl_cpws=Label(frm,text="Confirm password",font=("arial",20),bg="powder blue")
    lbl_cpws.place(relx=.22,rely=.7)
    
    e_cpws=Entry(frm,font=("arial",20),bd=2,width=16,show="*")
    e_cpws.place(relx=.5,rely=.7)

    bt_open=Button(frm,text="open",font=("arial",20),command=openacdb)
    bt_open.place(relx=.5,rely=.8)

    bt_back=Button(frm,text="Back",font=("arial",20),command=back)
    bt_back.place(relx=0,rely=0)

    
def fgscreen():
    
    frm=Frame(win)
    frm.configure(bg="powder blue")
    frm.place(relx=0,rely=.1,relwidth=1,relheight=.9)

    def back():
        frm.destroy()
        mainscreen()
        
    def reset():
        e_acn.delete(0,"end")
        e_pws.delete(0,"end")
        e_acn.focus()
    
    def recfgdb():
        
        name=e_name.get()
        email=e_eml.get()
        mob=e_ph.get()
        acn=e_acn.get()
        
        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select pswd from accounts where acn=? and name=? and mob=? and email=?",(acn,name,mob,email))
        tup=curobj.fetchone()
        if tup==None:
            messagebox.showerror("forgot password","Account does not exist")
        else:
            messagebox.showinfo("forgot password",f"your account password is {tup[0]}")
    
    lbl_name=Label(frm,text="Name",font=("arial",20),bg="powder blue")
    lbl_name.place(relx=.33,rely=.2)
    
    e_name=Entry(frm,font=("arial",20),bd=2,width=16)
    e_name.place(relx=.5,rely=.2)
                       
    lbl_acn=Label(frm,text="ACC no.",font=("arial",20),bg="powder blue")
    lbl_acn.place(relx=.33,rely=.3)
    
    e_acn=Entry(frm,font=("arial",20),bd=2,width=16)
    e_acn.place(relx=.5,rely=.3)
    
    lbl_eml=Label(frm,text="Email",font=("arial",20),bg="powder blue")
    lbl_eml.place(relx=.33,rely=.4)
    
    e_eml=Entry(frm,font=("arial",20),bd=2,width=16)
    e_eml.place(relx=.5,rely=.4)
    
            
    lbl_ph=Label(frm,text="mobile No.",font=("arial",20),bg="powder blue")
    lbl_ph.place(relx=.29,rely=.5)
    
    e_ph=Entry(frm,font=("arial",20),bd=2,width=16)
    e_ph.place(relx=.5,rely=.5)

    bt_back=Button(frm,text="Back",font=("arial",20),command=back)
    bt_back.place(relx=0,rely=0)
    
    bt_recover=Button(frm,text="Recover",font=("arial",20),command=recfgdb)
    bt_recover.place(relx=.5,rely=.6)
    
    bt_reset=Button(frm,text="reset",font=("arial",20),command=reset)
    bt_reset.place(relx=.6,rely=.6)
    
    
def queryscreen():
    
    frm=Frame(win)
    frm.configure(bg="powder blue")
    frm.place(relx=0,rely=.1,relwidth=1,relheight=.9)
    
    def back():
        frm.destroy()
        mainscreen()
        
    def sent():
        query=e_qt.get()
        acn=e_acn.get()
        name=e_name.get()
        
        if (len(query)==0 and len(acn)==0):
            messagebox.showerror("query","empty fields")
         
        else:
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select * from accounts where acn=? ",(acn,))
            tup=curobj.fetchone()
            
            if tup==None:
                messagebox.showerror("failed","acc does not found")
                
            else:
                conobj=sqlite3.connect(database="banking.sqlite")
                curobj=conobj.cursor()
                curobj.execute("insert into query values(?,?,?)",(acn,name,query))
                conobj.commit()
                conobj.close()
                messagebox.showinfo("Query sent","we help you ASAP")
            
    
    lbl_qt=Label(frm,text="How Can We Help You?",font=("arial",40),bg="powder blue")
    lbl_qt.pack()
    
    e_qt=Entry(frm,font=("arial",20),bd=2,width=50)
    e_qt.place(relx=.2,rely=.3)
    
    lbl_acn=Label(frm,text="Account no.",font=("arial",20),bg="powder blue")
    lbl_acn.place(relx=.2,rely=.4)
    
    e_acn=Entry(frm,font=("arial",20),bd=2,width=15)
    e_acn.place(relx=.4,rely=.4)
    
    lbl_name=Label(frm,text="Name",font=("arial",20),bg="powder blue")
    lbl_name.place(relx=.2,rely=.5)
    
    e_name=Entry(frm,font=("arial",20),bd=2,width=15)
    e_name.place(relx=.4,rely=.5)
    
    bt_sent=Button(frm,text="sent",font=("arial",20),command=sent)
    bt_sent.place(relx=.5,rely=.7)
    
    bt_back=Button(frm,text="Back",font=("arial",20),command=back)
    bt_back.place(relx=0,rely=0)
    
    
def homescreen():
    frm=Frame(win)
    frm.configure(bg="powder blue")
    frm.place(relx=0,rely=.1,relwidth=1,relheight=.9)
    
    
    def logout():
        frm.destroy()
        mainscreen()
    
    def detailscreen():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.configure(bg="white")
        ifrm.place(relx=.3,rely=.2,relwidth=.6,relheight=.7)
        
        def back():
            ifrm.destroy()
            homescreen()
        
        lbl_wel=Label(ifrm,text="This is Detail Screen",font=("arial",15,"bold"),bg="white")
        lbl_wel.pack()
        
        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from accounts where acn=?",(uacn,))
        tup=curobj.fetchone()
        
        lbl_acn=Label(ifrm,text=f"Account No.:\t {tup[0]}",font=("arial",15,"bold"),bg="white")
        lbl_acn.place(relx=.2,rely=.2)
        
        lbl_bal=Label(ifrm,text=f"balance :\t\t {tup[6]}",font=("arial",15,"bold"),bg="white")
        lbl_bal.place(relx=.2,rely=.3)
        
        lbl_opdate=Label(ifrm,text=f"open date :\t {tup[7]}",font=("arial",15,"bold"),bg="white")
        lbl_opdate.place(relx=.2,rely=.4)
        
        bt_back=Button(ifrm,text="Back",font=("arial",10),command=back)
        bt_back.place(relx=0,rely=0)
    
    def deposit():
        ifrm=Frame(frm)
        ifrm.configure(bg="white")
        ifrm.place(relx=.3,rely=.2,relwidth=.6,relheight=.7)
        
        def back():
            ifrm.destroy()
            homescreen()
        
        def depdb():
            amt=float(e_amt.get())
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select bal from accounts where acn=?",(uacn,))
            bal=curobj.fetchone()[0]
            curobj.close()
            curobj=conobj.cursor()
            curobj.execute("update accounts set bal=bal+? where acn=?",(amt,uacn))
            curobj.execute("insert into txns values(?,?,?,?,?)",(uacn,amt,bal+amt,"cr",time.ctime()))
            conobj.commit()
            conobj.close()
            messagebox.showinfo("credit",f"amount {amt} deposited succesfully")
            
        
        lbl_dep=Label(ifrm,text="This is Deposit Screen",font=("arial",15,"bold"),bg="white")
        lbl_dep.pack()
        
        lbl_amt=Label(ifrm,text="Amount",font=("arial",15,"bold"),bg="white")
        lbl_amt.place(relx=.3,rely=.2)
        
        e_amt=Entry(ifrm,font=("arial",20),bd=2,width=15)
        e_amt.place(relx=.4,rely=.2)
        
        bt_dep=Button(ifrm,text="deposit",font=("arial",18),command=depdb)
        bt_dep.place(relx=.45,rely=.3)
    
        bt_back=Button(ifrm,text="Back",font=("arial",10),command=back)
        bt_back.place(relx=0,rely=0)
        
    def withdraw():
        ifrm=Frame(frm)
        ifrm.configure(bg="white")
        ifrm.place(relx=.3,rely=.2,relwidth=.6,relheight=.7)
        
        def back():
            ifrm.destroy()
            homescreen()
            
        def wddb():
            amt=float(e_amt.get())
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select bal from accounts where acn=?",(uacn,))
            bal=curobj.fetchone()[0]
            curobj.close()
            if bal>amt:
                curobj=conobj.cursor()
                curobj.execute("update accounts set bal=bal-? where acn=?",(amt,uacn))
                curobj.execute("insert into txns values(?,?,?,?,?)",(uacn,amt,bal-amt,"db",time.ctime()))
                conobj.commit()
                messagebox.showinfo("credit",f"amount {amt} withdrawn succesfully")
            else:
                messagebox.showwarning("warning","insufficient balance")
            conobj.close()
        
        lbl_wel=Label(ifrm,text="This is Withdraw Screen",font=("arial",15,"bold"),bg="white")
        lbl_wel.pack()
        
        lbl_amt=Label(ifrm,text="Amount",font=("arial",15,"bold"),bg="white")
        lbl_amt.place(relx=.3,rely=.2)
        
        e_amt=Entry(ifrm,font=("arial",20),bd=2,width=15)
        e_amt.place(relx=.4,rely=.2)
        
        bt_wd=Button(ifrm,text="withdraw",font=("arial",18),command=wddb)
        bt_wd.place(relx=.45,rely=.3)
        
        bt_back=Button(ifrm,text="Back",font=("arial",10),command=back)
        bt_back.place(relx=0,rely=0)
        
    def trans():
        ifrm=Frame(frm)
        ifrm.configure(bg="white")
        ifrm.place(relx=.3,rely=.2,relwidth=.6,relheight=.7)
        
        def back():
            ifrm.destroy()
            homescreen()
            
        def transfer():
            to=e_to.get()
            amt=float(e_amt.get())
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select * from accounts where acn=?",(to,))
            tup=curobj.fetchone()
            curobj.close()
            if tup==None:
                messagebox.showerror("not found",f"{to} account does not fond")
            else:
                curobj=conobj.cursor()
                curobj.execute("select bal from accounts where acn=?",(uacn,))
                bal=curobj.fetchone()[0]
                curobj.close()
                if bal>=amt:
                    curobj=conobj.cursor()
                    curobj.execute("update accounts set bal=bal+? where acn=?",(amt,to))
                    curobj.execute("update accounts set bal=bal-? where acn=?",(amt,uacn))
                    curobj.execute("insert into txns values(?,?,?,?,?)",(to,amt,bal+amt,"cr",time.ctime()))
                    curobj.execute("insert into txns values(?,?,?,?,?)",(uacn,amt,bal-amt,"db",time.ctime()))
                    conobj.commit()
                    messagebox.showinfo("transfered",f"{amt} is transfer to acc no. {to}")
                else:
                    messagebox.showinfo("balance","insufficient balance")
                    conobj.close()
            
        
        lbl_wel=Label(ifrm,text="This is Transfer Screen",font=("arial",15,"bold"),bg="white")
        lbl_wel.pack()
        
        lbl_to=Label(ifrm,text="To Account no.",font=("arial",20),bg="white")
        lbl_to.place(relx=.2,rely=.4)
    
        e_to=Entry(ifrm,font=("arial",20),bd=2,width=15)
        e_to.place(relx=.5,rely=.4)
    
        lbl_amt=Label(ifrm,text="Amount",font=("arial",20),bg="white")
        lbl_amt.place(relx=.2,rely=.5)
    
        e_amt=Entry(ifrm,font=("arial",20),bd=2,width=15)
        e_amt.place(relx=.5,rely=.5)
        
        bt_sent=Button(ifrm,text="Sent",font=("arial",18),command=transfer)
        bt_sent.place(relx=.4,rely=.6) 
        
        bt_back=Button(ifrm,text="Back",font=("arial",10),command=back)
        bt_back.place(relx=0,rely=0) 
        
    def update():
        ifrm=Frame(frm)
        ifrm.configure(bg="white")
        ifrm.place(relx=.3,rely=.2,relwidth=.6,relheight=.7)
        
        def back():
            ifrm.destroy()
            homescreen()
        def updt():
            name=e_name.get()
            pswd=e_pws.get()
            email=e_eml.get()
            mob=e_ph.get()
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("update accounts set name=?,pswd=?,email=?,mob=? where acn=?",(name,pswd,email,mob,uacn))
            conobj.commit()
            conobj.close()
            messagebox.showinfo("Update","Profile Updated")
            ifrm.destroy()
            global uname
            uname=name
            update()
            
        
        
        lbl_wel=Label(ifrm,text="This is Update profile Screen",font=("arial",15,"bold"),bg="white")
        lbl_wel.pack()
        
        lbl_name=Label(ifrm,text="Name",font=("arial",20),bg="white")
        lbl_name.place(relx=.2,rely=.2)
    
        e_name=Entry(ifrm,font=("arial",20),bd=2,width=16)
        e_name.place(relx=.5,rely=.2)
    
        lbl_eml=Label(ifrm,text="Email",font=("arial",20),bg="white")
        lbl_eml.place(relx=.2,rely=.3)
    
        e_eml=Entry(ifrm,font=("arial",20),bd=2,width=16)
        e_eml.place(relx=.5,rely=.3)
    
            
        lbl_ph=Label(ifrm,text="mobile No.",font=("arial",20),bg="white")
        lbl_ph.place(relx=.2,rely=.4)
        
        e_ph=Entry(ifrm,font=("arial",20),bd=2,width=16)
        e_ph.place(relx=.5,rely=.4)
            
        lbl_pws=Label(ifrm,text="password",font=("arial",20),bg="white")
        lbl_pws.place(relx=.2,rely=.5)
    
        e_pws=Entry(ifrm,font=("arial",20),bd=2,width=16)
        e_pws.place(relx=.5,rely=.5)  
        
        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from accounts where acn=?",(uacn,))
        tup=curobj.fetchone()
        curobj.close()
        e_name.insert(0,tup[1])
        e_eml.insert(0,tup[2])
        e_ph.insert(0,tup[3])
        e_pws.insert(0,tup[4])
        
        
        bt_up=Button(ifrm,text="update",font=("arial",18),command=updt)
        bt_up.place(relx=.5,rely=.6)
        
        bt_back=Button(ifrm,text="Back",font=("arial",10),command=back)
        bt_back.place(relx=0,rely=0)
        
    def history():
        ifrm=Frame(frm)
        ifrm.configure(bg="white")
        ifrm.place(relx=.3,rely=.2,relwidth=.6,relheight=.7)
        
        def back():
            ifrm.destroy()
            homescreen()
        
        lbl_wel=Label(ifrm,text="This is History Screen",font=("arial",15,"bold"),bg="white")
        lbl_wel.pack()
        
        tv=Treeview(ifrm)
        tv.place(x=60,y=100,height=100,width=600)
        
        
        style = Style()
        style.configure("Treeview.Heading", font=('Arial',10,'bold'),foreground='brown')

        sb=Scrollbar(ifrm,orient='vertical',command=tv.yview)
        sb.place(x=650,y=100,height=100)
        tv.configure(yscrollcommand=sb.set)
        
        tv['columns']=('date','amt','type','upbal')
        
        tv.column('date',width=80,anchor='c')
        tv.column('amt',width=40,anchor='c')
        tv.column('type',width=40,anchor='c')
        tv.column('upbal',width=40,anchor='c')
        
        tv.heading('date',text='Txn Date')
        tv.heading('amt',text='Amount')
        tv.heading('type',text='Txn type')
        tv.heading('upbal',text='Updated Bal')
        
        tv['show']='headings'
        
        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from txns where acn=?",(uacn,))
        
        for row in curobj:
            tv.insert("","end",values=(row[4],row[1],row[3],row[2]))
        
        
        bt_back=Button(ifrm,text="Back",font=("arial",10),command=back)
        bt_back.place(relx=0,rely=0)
    
    lbl_wel=Label(frm,text=f"Welcome,{uname}",font=("arial",15,"bold"),bg="powder blue")
    lbl_wel.place(relx=.04,rely=0)
    
    bt_detail=Button(frm,text="Details",font=("arial",20),width=16,command=detailscreen)
    bt_detail.place(relx=0,rely=.2)

    bt_dep=Button(frm,text="Deposit",font=("arial",20),width=16,command=deposit)
    bt_dep.place(relx=0,rely=.3)
    
    bt_wd=Button(frm,text="Withdraw",font=("arial",20),width=16,command=withdraw)
    bt_wd.place(relx=0,rely=.4)

    bt_txn=Button(frm,text="Transfer",font=("arial",20),width=16,command=trans)
    bt_txn.place(relx=0,rely=.5)
    
    bt_his=Button(frm,text="History",font=("arial",20),width=16,command=history)
    bt_his.place(relx=0,rely=.6)
    
    bt_up=Button(frm,text="Update Profile",font=("arial",20),width=16,command=update)
    bt_up.place(relx=0,rely=.7)

    bt_log=Button(frm,text="Logout",font=("arial",15),command=logout)
    bt_log.place(relx=.94,rely=.0)
mainscreen()
win.mainloop()
```
